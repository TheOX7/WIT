```python
df.info()
```

```python
df['Pendapatan'] = [fake.random_int(min=1500000, max=3000000, step=150000) for _ in range(24)]
df['Umur'] = [fake.random_int(min=19, max=21) for _ in range (24)]
df['Asuransi'] = [fake.random_element(elements=('Ya', 'Tidak')) for _ in range(24)]
```

```python
export_file = 'WIT_2.xlsx'
df.to_excel(export_file, index=False)
```
